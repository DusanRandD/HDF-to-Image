```python
# Return dictionary with numpy array and print what file contain
from pyhdf.SD import SD, SDC
def hdf_to_numpy(hdf_file_path):
    hdf = SD(hdf_file_path, SDC.READ)
    hdf_info = list(hdf.datasets())
    data=[]
    print('HDF file contain:')
    for data_name in hdf_info:
        data.append(hdf.select(data_name)[:,:])
        print(data_name)
    return dict(zip(hdf_info,data))
```

```python
# Convert HDF file from given path to dictionary
HDF_filepath = '/kaggle/input/hdfile/'+filenames[0]
converted_data = hdf_to_numpy(HDF_filepath)
```

```python
# Plot and save figure function
import matplotlib.pyplot as plt
def print_image_form_dict(file_number,nan_black=True):
    HDF_filepath = '/kaggle/input/hdfile/'+filenames[file_number]
    converted_data = hdf_to_numpy(HDF_filepath)
    data = converted_data['Aerosol_Cloud_Fraction_Land']
    if nan_black:
        data[data==-9999]=0
    else:
        data[data==-9999]=1000
    image_np = data/1000
    plt.figure(figsize=(8,15))
    plt.axis('off')
    plt.imshow(image_np,cmap='gray')
    plt.savefig('Aerosol_Cloud_Fraction_Land_{}.png'.format(file_number))
```

```python
# Plot and save Image with given file number
print_image_form_dict(0)
```
